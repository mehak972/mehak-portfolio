```python
#Join census tracts and crimes, for spatial join, it has to be in same coordinate system
if df_23.crs != census_tracts.crs:
    df_23 = df_23.to_crs(census_tracts.crs) #To check if crs is same
    
crimes_2023=gpd.sjoin(
    df_23,census_tracts,
    how="left", #keep all pois
    predicate="within" #point within polygon
)

crimes_2023.head()


#df_23_assault=df_23[df_23[""]==""]
#df_23_assault["METHOD"].value_counts()

#print(list(crime_2023.columns))
```

```python
#Join census tracts and crimes
crimes_2024=gpd.sjoin(
    df_24,census_tracts,
    how="left", #keep all pois
    predicate="within" #point within polygon
)

crimes_2024.head()
```

```python
#Group by Census tract so each crime alligns with census tracts
num_23=(
    crimes_2023.groupby("TRACT").size()
    .reset_index(name="num_crimes2023")
    
)
num_23
```

```python
#Group by Census tract so each crime alligns with census tracts
num_24=(
    crimes_2024.groupby("TRACT").size()
    .reset_index(name="num_crimes2024")
    
)
num_24
```

```python
ct_counts=census_tracts.merge(num_23,on= "TRACT", how="left")
ct_counts = ct_counts[['TRACT', 'GEOID', 'OBJECTID', 'geometry', 'num_crimes2023']]
ct_counts
```

```python
ct_counts_23_24 = ct_counts.merge(num_24, on= "TRACT", how="left")
ct_counts_23_24
```

```python
ct_counts_23_24['change_24_23'] = ct_counts_23_24['num_crimes2024'] - ct_counts_23_24['num_crimes2023'] 

ct_counts_23_24
```

```python
ct_counts_23_24.crs
```

```python
# Reproject to UTM (e.g., Zone 18N, EPSG:32618; adjust for your location)
ct_counts_23_24 = ct_counts_23_24.to_crs(epsg=32618)
```

```python
# Adding North and scale bars
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm

def add_north_arrow(ax, xy=(0.90, 0.95), text="North"):
    """
    Add a simple north arrow. xy is in axes fraction coordinates (0..1).
    """
    ax.annotate(
        text, xy=xy, xytext=(xy[0], xy[1]-0.12),
        xycoords='axes fraction', textcoords='axes fraction',
        ha='center', va='center', fontsize=14, fontweight='bold',
        arrowprops=dict(arrowstyle='fancy', linewidth=2, color='black')
    )

def add_scale_bar(ax, length_m=1000, label="1 km", location='lower left'):
    """
    Add a simple scale bar using AnchoredSizeBar. CRS must be meters.
    length_m: length of the scale bar in meters.
    label: label text shown below the bar.
    """
    scalebar = AnchoredSizeBar(
        ax.transData, length_m, label, location,
        pad=1.0, color='black', frameon=False,
        size_vertical=length_m * 0.06,  # thickness of the bar
        fontproperties=fm.FontProperties(size=10)
    )
    ax.add_artist(scalebar)
```

```python
# Verify GeoDataFrame
print(ct_counts_23_24.columns)  # Confirm 'change_24_23' exists
print(ct_counts_23_24.crs)     # Should show EPSG:32618
```

```python
ax = ct_counts_23_24.plot(
    column="change_24_23", 
    cmap="coolwarm", #Try Blues, Grays, Greens, Oranges, Reds, or contrasting variations such as viridis, plasma, coolwarm, magma, inferno, cividis
    legend=True,
    edgecolor="black",
    linewidth=0.5,
    figsize=(12,12)
)
ax.set_title("Total crimes in 2024")
#ax.set_axis_off()

add_north_arrow(ax)
add_scale_bar(ax, length_m=2000, label="2 km", location='lower left')
```

```python
import contextily as cx
import matplotlib.pyplot as plt
# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Plot choropleth map with semi-transparent fill for basemap visibility
ct_counts_23_24.plot(
    column='change_24_23',  
    cmap='Greys',
    legend=True,
    edgecolor='black',
    linewidth=0.5,
    alpha=0.6,  # Semi-transparent to see basemap
    ax=ax,
    legend_kwds={'label': 'Change 2024-2023', 'orientation': 'vertical'}
)

# Add basemap (Contextily uses Web Mercator, EPSG:3857)
# Reproject data to EPSG:3857 for basemap alignment
cx.add_basemap(
    ax,
    crs=ct_counts_23_24.crs.to_string(),  # Match GeoDataFrame CRS (EPSG:32618)
    source=cx.providers.OpenStreetMap.DE,  # Use OpenStreetMap basemap  ### for different basemaps, visit: https://contextily.readthedocs.io/en/latest/intro_guide.html 
    #source=cx.providers.CartoDB.Positron,
    zoom='auto'
)

# Set title and add north arrow and scale bar
ax.set_title('Crime Change (2024 vs 2023) by Census Tract')
add_north_arrow(ax)
add_scale_bar(ax, length_m=2000, label='2 km')
ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# Count crimes by TRACT and OFFENSE
offense_counts = (
    crimes_2023.groupby(['TRACT', 'OFFENSE']).size()
    .reset_index(name='count')
    .pivot_table(index='TRACT', columns='OFFENSE', values='count', fill_value=0)
    .reset_index()
)
# Rename columns to replace special characters (e.g., '/' to '_')
offense_counts.columns = [col.replace('/', '_').replace(' ', '_') for col in offense_counts.columns]

offense_counts
```

```python
ct_offense_23 = census_tracts.merge(offense_counts, on= "TRACT", how="left")

ct_offense_23
```

```python
# Reproject to UTM (e.g., Zone 18N, EPSG:32618; adjust for your location)
ct_offense_23 =  ct_offense_23.to_crs(epsg=32618)
```

```python
## Plotting one type of crime offense 
ax = ct_offense_23.plot(
    column="HOMICIDE", 
    cmap="inferno", #Try Blues, Grays, Greens, Oranges, Reds, or contrasting variations such as viridis, plasma, coolwarm, magma, inferno, cividis
    legend=True,
    edgecolor="black",
    linewidth=0.5,
    figsize=(12,12)
)
ax.set_title("Homicides in DC in the year 2023")
ax.set_axis_off()

add_north_arrow(ax)
add_scale_bar(ax, length_m=2000, label="2 km", location='lower left')
```
